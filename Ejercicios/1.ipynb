Una de las principales ventajas de la programacin es la posibilidad de crear bucles y repeticiones para tareas especficas, y que no tengamos que realizar el mismo cdigo varias veces de forma manual.

Los bucles nos permiten simplificar nuestro cdigo, que sea ms fcil de leer e incluso ms fcil de modificar y mantener.



# Condicin
Al igual que en los condicionales **if**, en los bucles se va a **evaluar una condicin** para saber si se debe seguir repetiendo el bucle o se debe finalizar. Habitualmente, lo que se suele hacer es establecer que si la condicin **es verdadera**, se vuelve a repetir el bucle. Por el contrario, si es falsa, se finaliza. Sin embargo, esta condicin puede variar dependiendo de la implementacin que le indique el programador.

# Iterar
Otro concepto que usaremos mucho dentro de un bucle es el concepto de **Iteracin**. Esto se refiere a **cada una de las repeticiones de un bucle**. Por ejemplo, si un bucle se repite como en el grfico inferior de este artculo (de 0 a 3*), se dice que hay 4 iteraciones.



 
***Mucho cuidado con los bucles, porque en programacin se suele empezar a contar en 0, 
por lo que esa  es la primera iteracin. Si el bucle va desde 0 a 3, hay 4 iteraciones.***

# Contador
Muchas veces, los bucles que creamos incorporan un contador, que no es ms que algo que ir guardando un nmero para **contar el nmero de repeticiones realizadas**, y as **finalizar** cuando se llegue a otro **nmero concreto**. Dicho contador hay que inicializarlo (crearlo y darle un valor) antes de comenzar el bucle.

# Incremento/Decremento
Al igual que tenemos un contador en un bucle, tambin debemos tener una parte donde hagamos un **incremento (o un decremento) de dicho contador**. **Si no** lo tuvieramos, el contador *no cambiara* y la condicin **siempre sera veradera**, por lo que sera imposible salir del bucle.(**Bucle Infinito**)

 
***Ten en cuenta que de producirse un bucle infinto, nuestro programa se quedar atascado y 
tendremos que forzar para finalizarlo. Ten siempre cuidado al crear un bucle para que no sea infinito.***

Un buen hbito de programacin cuando creamos bucles es pensar la forma de hacer que esos bucles siempre vayan desde un nmero inicial a un nmero final y terminen las repeticiones. De esta forma son predecibles y fciles de leer. Sin embargo, en algunas ocasiones nos puede interesar hacer interrupciones o saltos de iteraciones para conseguir algo ms especfico que con un bucle ntegro es ms complejo de conseguir.

# Saltar una iteracin  **continue**

Imagina un caso donde queremos saltarnos una iteracin concreta. Por ejemplo, queremos hacer un bucle que muestre los nmeros del 0 al 10, pero queremos que el nmero 5 se lo salte y no lo muestre, continuando con el resto.

Para eso podemos hacer uso de continue, que es una sentencia que al llegar a ella dentro de un bucle, el programa salta y abandona esa iteracin, volviendo al principio del bucle:

`````python
for (let i = 0; i < 11; i++) {
    if (i == 5) {
      continue;
    }
  
    console.log("Valor de i:", i);
  }
`````

# Interrumpir el bucle  **break**

De la misma forma que tenemos un continue que interrumpe la iteracin y vuelve al inicio a evaluar la condicin, tenemos un break que nos permite interrumpir el bucle y abandonarlo. Esto puede ser bastante til cuando queremos que se abandone el bucle por una condicin especial.

Por ejemplo, cambiemos el continue del ejemplo anterior por el break:

`````python
for (let i = 0; i < 11; i++) {
    if (i == 5) {
      break;
    }
  
    console.log("Valor de i:", i);
  }
`````


> *No obstante, recuerda que estas interrupciones slo deben usarse en casos muy especficos. Siempre es preferible que el bucle sea predecible y no tenga demasiadas situaciones excepcionales que interrumpan el flujo, ya que a la larga son difciles de modificar y mantener.*

# while

El bucle **WHILE** es una estructura de control que **repite** un bloque de cdigo mientras una condicin se evale como **verdadera**.

En lenguaje natural, el bucle WHILE significa,

***Mientras se cumpla esto -> haz esto***

En cada iteracin, se verifica la condicin:

+ Si es *true*, se ejecuta el bloque de cdigo
+ Si es *false*, el bucle se detiene y el programa contina con la siguiente instruccin despus del bucle

Si la condicin se evala como falsa desde el principio, el bloque de cdigo no se ejecutar en absoluto.

`````python
while(condicion)
{
    // Bloque de cdigo a ejecutar mientras la condicin sea verdadera
}

// resto de cdigo
`````

`````python
let i = 0;  // Inicializacin de la variable contador

// Condicin: Mientras la variable contador sea menor de 5
while (i < 5) {
  console.log("Valor de i:", i);

  i++ //(i = i + 1; Incrementamos el valor de i)
}

// Valor de i: 0
// Valor de i: 1
// Valor de i: 2
// Valor de i: 3
// Valor de i: 4
`````

En este ejemplo:

El bucle WHILE se ejecuta mientras la variable contador sea menor que 5. En cada iteracin realizaramos una accin, que podra emplear el valor de contador.

Despus de cada iteracin, se incrementa el contador en 1. Si no hiciramos esto, la condicin nunca se cumplira, y caeramos en un bucle infinito.

***La operacin i = i + 1 es lo que se suele llamar un incremento de una variable. Es muy comn simplificarla como i++, que hace exactamente lo mismo: aumenta en 1 su valor.***

# do..while

El bucle **do..while** es una estructura de control que repite un bloque de cdigo al menos una vez, y luego verifica una condicin para decidir si continuar o salir del bucle.

En lenguaje natural, el bucle WHILE significa,

***Haz esto -> mientras se cumpla esto***

El bucle *do..while* es el hermano del bucle *while*, y su funcionamiento es muy parecido.

La diferencia es que en el bucle do..while la condicin se verifica despus de ejecutar el cuerpo del bucle. Por tanto, el bucle se ejecuta al menos una vez.

**El bloque de cdigo se ejecuta primero y luego se verifica la condicin**. Si la condicin se evala como *verdadera*, el bucle se repite y el bloque de cdigo se ejecuta nuevamente. Si la condicin se evala como *falsa, el bucle se detiene* y el programa contina con la siguiente instruccin despus del bucle.




`````python
do
{
    // Bloque de cdigo a ejecutar mientras la condicin sea verdadera
}
while(condicion)

// resto de cdigo
`````

`````python
let i = 5;

do {
  console.log("Hola a todos");
  i++; // i = i + 1;
} while (i < 5);

console.log("Bucle finalizado");
`````

# for

El bucle **for** es una estructura de control que permite **repetir un bloque de cdigo un nmero especfico de veces**. Es una de las estructuras de control ms utilizadas y poderosas.

Ya hemos visto los bucles WHILE y DO-WHILE, que son los bucles ms bsicos.

A medida que se usaban los bucles, ocurra con mucha frecuencia que se tena la necesidad de hacer algo N veces (100, 1.000, 1.000.000). Sin depender de una condicin, simplemente

**Quiero hacer esto 100 veces...**

Eso se poda hacer con un bucle WHILE de forma sencilla, claro pero...

Por ejemplo, si quisiera hacer una accin 100 veces podramos usar un contador y hacer:

````
contador = 0

while(contador < 100)
{
    // hacer algo
    contador += 1;
}
````
Pero esta estructura se repeta tanto, tanto, que  eh, y si.....

si hacemos algo para que sea ms sencillo?. Y as surge el bucle FOR.

`````python
for(inicializacion; condition; incremento)
{
	// hacer algo
}
`````

En este bucle FOR, tal cul est definido en muchos lenguajes, consta de tres partes.

+ ***inicializacin***; una sentencia que se ejecuta antes de entrar en el bucle  
    ***creamos una variable temporal llamada i***

+ ***condition***; sentencia que se evala para continuar el bucle  
    ***realizamos el bucle mientras i < 100***

+ ***incremento*** sentencia que se ejecuta en cada iteracin del bucle  
    ***incrementamos i con i++***

`````python
for (let i = 0; i < 100; i++)
{
    // hacer algo 
}
`````

> *En principio, el programador debera realizar el bucle con el que ms cmodo se sienta, tanto un WHILE como un FOR. Sin embargo, es ms comn encontrarse con bucles for en el da a da, por lo que se recomienda no dejar de utilizar el for simplemente por un rechazo inicial debido a que su sintaxis pueda parecer ms compleja.*

# for ...  of

Los bucles **for in** y **for of** en JavaScript son dos modos de escribir el bucle for en este lenguaje de programacin. Este **tipo de bucle** es muy comn en el desarrollo web y normalmente se utiliza para **recorrer los distintos elementos de un array**.

El bucle **for of** nos facilita la escritura de nuestro bucle al *tener un contador interno*. Esto hace que se reduzcan los elementos que debemos definir dentro de nuestro bucle. Entonces, el bucle for of est compuesto por **dos elementos:**:  
Una **variable temporal** en la que tenemos el valor actual dentro del loop y un **array a recorrer**. La variable temporal es una forma de almacenar el valor de cada ciclo del loop.

`````python
let elementos = ["Agua","Tierra","Fuego","Aire"];

for (elemento of elementos){
console.log(elemento);
}

//Imprime los valores Agua, Tierra, Fuego y Aire
`````

`````python
// for of
let languagesArray = [
  "Spanish",
  "Euskera",
  "English",
  "Italian",
  "German",
  "French",
  "Maths",
];

for (let Language of languagesArray) {
  console.log(`Language is ${Language}`);
}

// salida

/* Language is Spanish
Language is Euskera
Language is English
Language is Italian
Language is German
Language is French
Language is Maths
 */
`````

> *Como el bucle tiene un contador interno, no tenemos ningn control sobre el ndice de los elementos. haciendo que todos sean visibles en el orden en el que los hayamos escrito en el array.*

# for ... in

Un bucle **for in** lo constituyen igualmente **dos elementos**: un **ndice temporal actual** dentro del loop y un **array a recorrer**, que llamamos con su nombre. Estos dos elementos los separa la palabra clave in. A continuacin, te presentamos cmo se escribira:

Este tipo de for, usado de forma "simple" solo va a imprimir los ndices del array que se le asigne.

`````python
let elementos = ["Agua","Tierra","Fuego","Aire"];

for (elemento in elementos){
console.log(elemento);
}

//Imprime los ndices 0,1,2,3
`````

Aunque se puede jugar con los datos dentro de este ciclo para imprimir los valores que contenga el array y no sus ndices.

`````python
let elementos = ["Agua","Tierra","Fuego","Aire"];

for (elemento in elementos){
console.log(elementos[elemento]);
}

//Imprime los valores Agua, Tierra, Fuego y Aire
`````

> Es muy parecido a hacer algo como elementos [ i ] en un for loop normal

````
console.log (elemento [i]);
````
En nuestro ejemplo....

`````python
let languagesArray = [
  "Spanish",
  "Euskera",
  "English",
  "Italian",
  "German",
  "French",
  "Maths",
];

for (let index in languagesArray) {
  let language = languagesArray[index];
  console.log(`Language in position ${index} is ${language}`);
}

// salida

/* Language in position 0 is Spanish
Language in position 1 is Euskera
Language in position 2 is English
Language in position 3 is Italian
Language in position 4 is German
Language in position 5 is French
Language in position 6 is Maths */
`````

# forEach

Este mtodo entra dentro del grupo de Iterables sin devolver una nueva matriz, lo que hace el **forEach** es ejecutar una funcin por cada elemento del array. En cada iteracin se tendr acceso a 3 variables: **valor** (del elemento), **ndice** (del elemento) y **array** (que estamos recorriendo) y con ello la funcin que se ejecutar para cada elemento de una matriz.

Este mtodo es muy til cuando solo necesitamos ejecutar una funcin a travs de cada elemento del array, sin necesidad de obtener un retorno. El forEach() tambin se puede utilizar en mapas y conjuntos.

> ***Tambin denominado:  
forEach()mtodo***

`````python
//  forEach()mtodo es:

array.forEach(function callback(valor, posicion, arrayUsado) {      
  //instrucciones a ejecutar
});
`````

`````python
let students = ['John', 'Sara', 'Jack'];

// using forEach
students.forEach(myFunction);

function myFunction(item) {

    console.log(item);
}

// salida
/* John
   Sara
   Jack */
`````

> el *forEach()* *mtodo* toma *myFunction()* una funcin que muestra cada elemento de estudiantes
Es bastante sencillo actualizar los elementos de la matriz si los necesitamos

`````python
let students = ['John', 'Sara', 'Jack'];

// using forEach
students.forEach(myFunction);

function myFunction(item, index, arr) {

    // adding strings to the array elements
    arr[index] = 'Hello ' + item;
}

console.log(students);

// salida
// ["Hola John", "Hola Sara", "Hola Jack"]
`````

#### Pero quin es ms rpido? Qu opcin es la ms aconsejable?  
#### ***for***    ***forEach*** 

Si tenemos que comparar en trminos de rendimiento ambos bucles, la cosa se pone un poco complicada. La razn de fondo es que la mayora de los navegadores modernos incluyen mtodos que detectan ambos y optimizan el resultado. Sin embargo, es posible hacernos una idea de quin es el *pistolero ms rpido al oeste del ro JavaScript* enumerando paso a paso los pasos que realiza cada ciclo for y foreach:

* Pasos a realizar por el ciclo ***for***:

    + 1. Establecer el valor inicial de la variable de control
    + 2. Comprobar si se debe o no salir del ciclo
    + 3. Ejecutar el cuerpo del ciclo
    + 4. Incrementar la variable de control  
     * Volver al paso 2*


* Mientras ***forEach***:

    + 1. Instanciar la funcin de llamada del ciclo
    + 2. Comprobar si hay un siguiente elemento a procesar
    + 3. Llamar a la funcin nuevamente para el siguiente elemento a procesar, con un nuevo contexto de ejecucin (esto comprende el alcance de la funcin; por lo tanto, su contexto, argumentos, variables internas y referencias a cualquier variable externa  si se utilizan).
    + 4. Ejecutar el contenido de la funcin
    + 5. Finalizacin de la llamada a la funcin  
     * Volver al paso 2*

*Podemos deducir que un bucle *for* es mucho ms rpido en tiempo de ejecucin que un bucle *forEach*. Pero hay que tener en cuenta que un bucle forEach imprime slo aquellas posiciones de un array que contengan informacin omitiendo posiciones vacas. Esto hace que un array complejo forEach se muestre ms eficaz a la hora de manejar nuestro cdigo a cambio de sacrificar rendimiento.*

En trminos prcticos, si comparamos el *for* con el *forEach*, podemos ver adems que es ms cmodo y limpio usar el forEach, ya que slo invocamos el mtodo foreach tambin especificamos una funcin de devolucin, donde va a estar iterando y devolviendo cada uno de los elementos y cuando termina, simplemente contina con la siguiente instruccin.
